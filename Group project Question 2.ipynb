```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define levels of predictors
social_time_levels = ["0", "1-2", "more than 2"]
gender_levels = ["other", "woman"]
ages = np.linspace(20, 80, 10)  # Example age range from 20 to 80

# Example coefficients from your table
coefficients = {
    "main_effect_age": 0.03,
    "social_time_0_gender_other": -1.5891,
    "social_time_more_gender_other": -1.8732,
    "social_time_0_gender_woman": -0.0577,
    "social_time_more_gender_woman": -0.9104,
    "age_social_time_more": -0.0285,
    "age_social_time_more_gender_other": 0.0284,
    "age_social_time_more_gender_woman": 0.0203
}

# Create a DataFrame for predictions
data = []
for age in ages:
    for gender in gender_levels:
        for social_time in social_time_levels:
            # Calculate predicted effect based on coefficients
            effect = coefficients["main_effect_age"] * age
            if social_time == "0":
                effect += coefficients[f"social_time_0_gender_{gender}"]
            elif social_time == "more than 2":
                effect += coefficients["age_social_time_more"] * age
                effect += coefficients[f"social_time_more_gender_{gender}"]
                effect += coefficients[f"age_social_time_more_gender_{gender}"] * age
            
            data.append({
                "Age": age,
                "Gender": gender,
                "Social Time": social_time,
                "Effect": effect
            })

df = pd.DataFrame(data)

# Plot interaction effect
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=df,
    x="Age",
    y="Effect",
    hue="Gender",
    style="Social Time",
    markers=True
)
plt.title("Interaction Effect of Social Time, Gender, and Age on Outcome")
plt.xlabel("Age")
plt.ylabel("Predicted Effect")
plt.legend(title="Gender & Social Time")
plt.grid(True)
plt.show()
```
